# Decision Trees

Decision Tree algorithm belongs to the family of supervised learning algorithms. 
Unlike other supervised learning algorithms, decision tree algorithm can be used for solving **Regression** and **Classification** problems

The primary challenge in the Decision Tree implementation is to identify **which attributes do we need to consider as the root node and each level**. Handling this is know the Attributes Selection. 
We have different Attributes Selection measures to identify the root node at each level. Two most commonly used methods are:

 - Information Gain
 - Gini Index 
 
 
Let's take a basic example to understand the mathematics behind Gini and Information Gain

## Practical Examples

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('D://DataScienceCollection//DataSets//IRIS//iris.csv')
```

```python
data.head(5)
```

```python
features=data[['sepallength','sepalwidth','petallength','petalwidth']]
target=data[['class']]
```

### Using GINI :

```python
# Create decision tree classifier object
decisiontree = DecisionTreeClassifier(criterion='gini', random_state=0)

# Train model
model = decisiontree.fit(features, target)

# Make new observation
observation = [[ 5,  4,  3,  2]]

# Predict observation's class
model.predict(observation)
```

### Using IG :

```python
# Create decision tree classifier object
decisiontree = DecisionTreeClassifier(criterion='entropy', random_state=0)

# Train model
model = decisiontree.fit(features, target)

# Make new observation
observation = [[ 5,  4,  3,  2]]

# Predict observation's class
model.predict(observation)
```

### Visualizing

```python
from IPython.display import Image  
from sklearn import tree
import pydotplus
```

```python
target_names=['setosa', 'versicolor' ,'virginica']
feature_names=['sepal length', 'sepal width', 'petal length', 'petal width']
```

```python
# Create DOT data
dot_data = tree.export_graphviz(decisiontree,
                                out_file=None,
                                feature_names=feature_names,
                                class_names=target_names)
```

```python
# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)

# Show graph
Image(graph.create_png())
```
