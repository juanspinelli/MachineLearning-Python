```python
import spacy
import string
import pandas as pd
from spacy.lang.en.stop_words import STOP_WORDS
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
nlp = spacy.load('en')
```

### Convert sentence into Lemma -----> Removal of Pronoun -----> Removal of  Stop Words ----> Removal of Punctuations

```python
#docx = nlp("This is how John Walker was walking. He was also running beside the lawn.")
docx = nlp(open('C://Users//Lenovo//Desktop//IPY//Data//Data//nlpPracticeText.txt').read())
dataAfterLemmaFilter = []
dataAfterPronounFilter = []
dataAfterStopwords =[]
dataAfterPunctuations =[]
dataAfterNounFilter =[]
```

### Removal of Lemmas

```python
for token in docx:
    dataAfterLemmaFilter.append(token.lemma_)
```

### Removal of Pronoun

```python
for token in dataAfterLemmaFilter:
    if token != "-PRON-":
        dataAfterPronounFilter.append(token.lower().strip())
```

### Removal of  Stop Words

```python
stopwords = list(STOP_WORDS)

for token in dataAfterPronounFilter:
    if token != stopwords:
        dataAfterStopwords.append(token) 
```

### Removal of Punctuations

```python
punctuations = string.punctuation

for token in dataAfterStopwords:
    if token not in punctuations:
        dataAfterPunctuations.append(token) 
```

### Selecting Nouns

```python
for value in dataAfterPunctuations:
    td = nlp(value)
    for t in td:
        if t.pos_ == 'NOUN':
            dataAfterNounFilter.append(t)
```

### Word count and Frequency

```python
# dataAfterNounFilter contains nlp objects not string. therefore, we are convertinf the nlp objects into string.
```

```python
dataAfterNounFilterStringFormatting =[]

for nlpObject in dataAfterNounFilter:
    nlpObjectIntoString=str(nlpObject)
    dataAfterNounFilterStringFormatting.append(nlpObjectIntoString)
```

```python
word_freq = Counter(dataAfterNounFilterStringFormatting)
commonWords = word_freq.most_common(2)
print(commonWords)
```

#### In case of confusion

```python
doubtfulWords = ['who']
for value in doubtfulWords:
    td = nlp(value)
    for t in td:
        print(t,"-------> ", t.pos_)
```

### Preparing Matrix

```python
docx = nlp(open('C://Users//Lenovo//Desktop//IPY//Data//Data//nlpPracticeText.txt').read())

sentence = str(docx)
doc = nlp(sentence)
text_data=[sent.string.strip() for sent in doc.sents]
print(len(text_data))

tfidf = TfidfVectorizer()
print(tfidf)
feature_matrix = tfidf.fit_transform(text_data)

tfidf.get_feature_names()

pd.DataFrame(feature_matrix.toarray(), columns=tfidf.get_feature_names())
```

### Preparing Matrix from Sentence -The same thing !!

```python
sentence = "I love Brazil, but I think Sweden is best. Not to mention that Germany beats both."
doc = nlp(sentence)
text_data=[sent.string.strip() for sent in doc.sents]
print(text_data)

count = CountVectorizer()
print(count)
bag_of_words = count.fit_transform(text_data)
print(bag_of_words)
feature_names = count.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns=feature_names)
```
