```python
girls = pd.read_csv('D:/DataScienceCollection/DataSets/schools_with_gtoilet.csv')
```

```python
girls
```

```python
print(list(girls.columns))
```

```python
girls_Pivot_State_UT_Year = pd.pivot_table(girls, index = ['State_UT','year'])
```

```python
girls_Pivot_State_UT_Year.head(15)
```

```python
girls_Pivot_State_UT_Year_Count = pd.pivot_table(girls, index = ['State_UT'],  aggfunc='count')
```

```python
girls_Pivot_State_UT_Year_Count.head(5)
```

```python
girls_Pivot_State_UT_Year_Median = pd.pivot_table(girls, index = ['State_UT'],  aggfunc='median')
```

```python
girls_Pivot_State_UT_Year_Median.head(5)
```

```python
girls_Pivot_State_UT_Year_Mean = pd.pivot_table(girls, index = ['State_UT'],  aggfunc='mean')
```

```python
girls_Pivot_State_UT_Year_Mean.head(5)
```

```python
girls_Pivot_State_UT_Year_AP = girls_Pivot_State_UT_Year.loc['Andhra Pradesh','Primary_Only']
```

```python
girls_Pivot_State_UT_Year_AP.plot(kind="bar",color=['r','b','g'])
plt.xlabel('Andhra Pradesh')
```
